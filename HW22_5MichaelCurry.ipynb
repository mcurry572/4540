This CoLab was created to help visual the optimization of J(θ) = (2θ-3)2
Step (i) is plotting J(0) vs 0 values by creating a range and then doing the calculations for that range.

Step (ii) is showcasing Gradient descent and goes through 10 iterations and plots them on a graph so that we can see what the minimum is.

You can change the values and run again to see what the optimal value for that equation is as well. If you change the iterations you will see that the value currently will never go under 1.5

```python
# prompt: 2.3. Perform the optimization problem of finding the minimum of J(θ) = (2θ-3)2 by:
# (i) plotting J(θ) vs θ.
# (ii) determining its minimum using gradient descent approach starting from a random initial value of θ = 5. Perform the search for the minimum using the gradient descent approach by hand calculations
# 2.5. Develop a Notebook tutorial to explain the process using the example 2.3 above, performing the optimization numerically. Do what 2.5 says only and use what is in 2.3

import numpy as np
import matplotlib.pyplot as plt

# 2.3 Optimization Problem

# (i) Plotting J(θ) vs θ
theta_values = np.linspace(-5, 10, 400)  # Create a range of theta values
J_theta_values = (2 * theta_values - 3)**2  # Calculate J(θ) for each theta

plt.figure(figsize=(8, 6))
plt.plot(theta_values, J_theta_values)
plt.xlabel("θ")
plt.ylabel("J(θ)")
plt.title("Plot of J(θ) vs θ")
plt.grid(True)
plt.show()


# (ii) Gradient Descent
def gradient_descent(initial_theta, learning_rate, iterations):
    theta = initial_theta
    theta_history = [theta]
    for i in range(iterations):
        gradient = 4 * (theta - 1.5)  # Derivative of J(θ)
        theta = theta - learning_rate * gradient
        theta_history.append(theta)
    return theta, theta_history


initial_theta = 5
learning_rate = 0.1
iterations = 20
optimal_theta, theta_history = gradient_descent(initial_theta, learning_rate, iterations)

print(f"Optimal theta: {optimal_theta}")

plt.figure(figsize=(8, 6))
plt.plot(range(iterations + 1), theta_history)
plt.xlabel("Iteration")
plt.ylabel("θ")
plt.title("Gradient Descent Progress")
plt.grid(True)

plt.show()


```
